# Company News Vs Stocks Researcher 

```python
import pandas as pd

# Generate a date range for 2023
start_date = "2019-01-01"
end_date = "2023-12-31"
dates_2023 = pd.date_range(start=start_date, end=end_date)
dates_2023
```

```python
import yfinance as yf

# Fetch stock data for HelloFresh for 2023
hellofresh_stock = yf.download('HFG.DE', start=start_date, end=end_date)

# Calculate daily percentage change
hellofresh_stock['Percent Change'] = hellofresh_stock['Adj Close'].pct_change() * 100

# Calculate rolling averages
hellofresh_stock['1d Avg Change'] = hellofresh_stock['Percent Change'].rolling(window=2).mean()
hellofresh_stock['5d Avg Change'] = hellofresh_stock['Percent Change'].rolling(window=6).mean()
hellofresh_stock
```

```python
threshold = 2
```

```python
threshold = 3
while True:
    significant_changes = hellofresh_stock[(hellofresh_stock['1d Avg Change'].abs() > threshold) | (hellofresh_stock['5d Avg Change'].abs() > threshold)]
    if significant_changes.shape[0]<= 30:
        break
    else :
        threshold += 0.02

significant_changes
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

def search_news(company, date):
    search_url = f"https://www.duckduckgo.com/html/?q={company}+news+after:{date}+before:{date}"
    response = requests.get(search_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    news_items = soup.find_all('a', class_='result__a')
    results = []
    for item in news_items:
        title = item.get_text()
        link = item['href']
        results.append({'title': title, 'link': link})
    
    return results

significant_changes['news'] = significant_changes.apply(lambda x: search_news('hello fresh canada ', str(x.index)), axis=1)
significant_changes
```

```python
significant_changes['news'].dropna(inplace=True)
for i in significant_changes['news']:
    if i:
        print(i)
```
