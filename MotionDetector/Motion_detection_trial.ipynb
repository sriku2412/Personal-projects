```python
# Thanks to https://github.com/misbah4064/motion_detection for the code
# thanks for your youtube tutorial - https://www.youtube.com/watch?v=_zKfYOriFMM 
# This code will detect motion in a video and its purely for educational purposes.

import cv2

background = cv2.imread("background2.png")
background = cv2.cvtColor(background, cv2.COLOR_BGR2GRAY)
background = cv2.GaussianBlur(background, (21, 21), 0)

video = cv2.VideoCapture("test3.avi")

while True:
    status, frame = video.read()

    if not status:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (21, 21), 0)
    
    diff = cv2.absdiff(background, gray)
    
    thresh = cv2.threshold(diff, 30, 255, cv2.THRESH_BINARY)[1]
    thresh = cv2.dilate(thresh, None, iterations=2)
    
    cnts, res = cv2.findContours(
        thresh.copy(),
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
        )
    
    for cnt in cnts:
        if cv2.contourArea(cnt) < 10000:
            continue
        (x, y, w, h) = cv2.boundingRect(cnt)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)
    
    cv2.imshow("Motion Detection", frame)
    # cv2.imshow("Threshold Video", thresh)
    # cv2.imshow("Difference video", diff)
    # cv2.imshow("Gray Video", gray)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
# testing with what i have learnt
background = cv2.imread("Screenshot 2024-02-09 170051.png")
background = cv2.cvtColor(background, cv2.COLOR_BGR2GRAY)
background = cv2.GaussianBlur(background, (21, 21), 0)

video = cv2.VideoCapture("20240209_165458.mp4")
status, frame = video.read()
# Resize the background to match frame size. 
background = cv2.resize(background, (frame.shape[1], frame.shape[0]))

while True:
    status, frame = video.read()

    if not status:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (21, 21), 0)

    diff = cv2.absdiff(background, gray)

    thresh = cv2.threshold(diff, 48, 255, cv2.THRESH_BINARY)[1]
    thresh = cv2.dilate(thresh, None, iterations=2)




    cnts, res = cv2.findContours(
        thresh.copy(),
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
        )

    for cnt in cnts:
        if cv2.contourArea(cnt) < 50:
            continue
        (x, y, w, h) = cv2.boundingRect(cnt)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)

    cv2.imshow("Motion Detection", frame)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
